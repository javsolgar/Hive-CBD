```python
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

# Limpieza de Google store

```python
data = pd.read_csv('./Google-Playstore.csv')
data.head()
```

## Limpieza de simbolo \ para poder utilizarlo como separador

```python
def busca_simbolo(data, simbolo, columnas):
    total = 0
    arreglar = []
    
    for columna in tqdm(columnas):
        indices = data[columna].str.find(simbolo)

        for posicion, existe in enumerate(indices):
            
            if existe != -1 and str(existe)!='nan':
                total +=1
                
                if columna not in arreglar:
                    arreglar.append(columna)
                    
    return total, arreglar
```

```python
columns = ['App Name', 'App Id', 'Category', 'Installs', 'Currency', 'Size', 'Minimum Android', 'Developer Id',
           'Developer Website', 'Developer Email', 'Released', 'Last Updated', 'Content Rating', 'Privacy Policy']

total, arreglar = busca_simbolo(data, '\\', columns)

print(total, arreglar)
```

```python
for column in arreglar: 
    data[column] = data[column].str.replace('\\','|')
```

```python
total, arreglar = busca_simbolo(data, '\\', columns)
print(total, arreglar)
```

## Limpieza de fecha

```python
data.head()
```

```python
def prepara_cadena(cadena):
    fechas = {
    'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06',
    'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12',
}
    if str(cadena) != 'nan':
        mes_dia, anyo = cadena.split(',')
        mes, dia = mes_dia.split(' ')
        mes = fechas[mes]
        return '{}/{}/{}'.format(dia, mes, anyo[1:])
    else:
        return cadena
    
def limpia_fecha(datos, columna):
    datos[columna] = datos[columna].apply(prepara_cadena)
    datos[columna] = pd.to_datetime(datos[columna], format='%d/%m/%Y')
    return datos
```

```python
data = limpia_fecha(data, 'Released')
data = limpia_fecha(data, 'Last Updated')
```

```python
data.head()
```

```python
data.to_csv('datos_limpiados.csv', sep='\\', index=False)
```

# Limpieza de Apple store

```python
data3 = pd.read_csv('365k_appstore.csv')
data3.head()
```

```python
columns = ['category','contentRating','currentVersion','description','logolink_large','logolink_medium','logolink_small','minimumOsVersion','name','price_currency','price_display','updatedAt','url']
total, arreglar = busca_simbolo(data3, '\\', columns)
print(total, arreglar)
```

```python
for column in arreglar:
    data3[column] = data3[column].str.replace('\\','|')
```

```python
total, arreglar = busca_simbolo(data3, '\\', columns)
print(total, arreglar)
```

```python
def prepara_cadena3(cadena):
    fecha = cadena.split('T')[0].strip()
    anyo, mes, dia = fecha.split('-')
    return '{}/{}/{}'.format(dia, mes, anyo)

def limpia_fecha3(datos, columna):
    datos[columna] = datos[columna].apply(prepara_cadena3)
    datos[columna] = pd.to_datetime(datos[columna], format='%d/%m/%Y')
    return datos
```

```python
data3 = limpia_fecha3(data3, 'updatedAt')
```

```python
data3.head()
```

```python
data3 = data3.drop('description', axis=1)
```

```python
data3.to_csv('datos3_limpiados.csv', sep='\\', index=False)
```
